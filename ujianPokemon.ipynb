```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
import pickle
```

```python
df = pd.read_csv('Pokemon.csv')
df
```

```python
df = df.dropna()
df.isnull().sum()
```

```python
df = df[['Name','Type 1','Generation','Legendary']]
df
```

```python

# mgenre = ext.fit_transform(df['Type 1'])
# print(ext.get_feature_names())
# print(len(ext.get_feature_names()))
```

```python
df['poke'] = df['Type 1'] + 'space' + df['Generation'].apply(str) + 'space' + df['Legendary'].apply(str)
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
ext = CountVectorizer(tokenizer = lambda x: x.split('space'), analyzer='word')
```

```python
matrix = ext.fit_transform(df['poke'])
pokemon = ext.get_feature_names()
countpokemon = len(pokemon)
dfmx = matrix.toarray()
print(pokemon)
print(countpokemon)
dfmx
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
score = cosine_similarity(matrix)
score
```

```python
like = 'Bulbasaur'
likeIndex = df[df['Name'] == like].index[0]
likeIndex
```

```python
allPokemon = list(enumerate(score[likeIndex]))
# allPokemon
```

```python
similarPokemon = []
for i in range(len(allPokemon)):
    if int(allPokemon[i][0]) != int(likeIndex):
        similarPokemon = (sorted(allPokemon,key = lambda x: x[1],reverse = True))
```

```python
listrecommend = []
for i in range(len(similarPokemon)):
    if int(similarPokemon[i][0]) != int(likeIndex):
        listrecommend.append(similarPokemon[i])
        
listrecommend[:6]
```

```python
pkl_filename = "pokemon_model.pkl"
with open(pkl_filename,'wb') as file:
    pickle.dump(score,file)
```
